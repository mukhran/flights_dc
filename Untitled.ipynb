```python
import os
import pandas as pd

folder_path = "./employment_airlines"
filenames = os.listdir(folder_path)

for filename in filenames:
    if filename.endswith(".xls"):
        filepath = os.path.join(folder_path, filename)

        try:
            # Read the Excel file skipping the header rows
            df = pd.read_excel(filepath, skiprows=3)

            # Specify the columns where you want to remove commas
            columns_to_clean = ['Full-time', 'Part-time', 'Grand Total']

            # Remove commas from specified columns
            df[columns_to_clean] = df[columns_to_clean].replace(",", "", regex=True)

            # Remove unwanted header line
            df = df[df['Month'].notna()]  # Assuming 'Month' column is always present in the valid data rows

            csv_filename = filename.replace(".xls", ".csv")
            df.to_csv(csv_filename, index=False)
            print(f"{filename} converted to CSV and cleaned.")

        except Exception as e:
            print(f"Error processing file {filename}: {str(e)}")
```

```python
import os
import pandas as pd

folder_path = "./employment_airlines"
filenames = os.listdir(folder_path)

for filename in filenames:
    if filename.endswith(".xls"):
        filepath = os.path.join(folder_path, filename)

        try:
            # Check file extension before attempting to read
            if not os.path.isfile(filepath):
                raise FileNotFoundError(f"File not found: {filepath}")

            # Read the Excel file skipping the header rows
            df = pd.read_excel(filepath, skiprows=3)

            # Specify the columns where you want to remove commas
            columns_to_clean = ['Full-time', 'Part-time', 'Grand Total']

            # Remove commas from specified columns
            df[columns_to_clean] = df[columns_to_clean].replace(",", "", regex=True)

            # Remove unwanted header line
            df = df[df['Month'].notna()]  # Assuming 'Month' column is always present in the valid data rows

            csv_filename = filename.replace(".xls", ".csv")
            df.to_csv(csv_filename, index=False)
            print(f"{filename} converted to CSV and cleaned.")

        except Exception as e:
            print(f"Error processing file {filename}: {str(e)}")
```

```python
import os
import pandas as pd

# Specify the path to your folder containing .xls files
folder_path = './employment_airlines'

# List all files in the folder
files = os.listdir(folder_path)

# Loop through each file
for file in files:
    if file.endswith('.xls'):
        # Construct the full file path
        file_path = os.path.join(folder_path, file)
        
        # Read the Excel file into a DataFrame
        df = pd.read_excel(file_path, skiprows=2)  # Skip the first row which contains the title
        
        # Remove commas from all columns
        df = df.applymap(lambda x: str(x).replace(',', ''))
        
        # Save the DataFrame to a CSV file
        csv_file_path = os.path.splitext(file_path)[0] + '.csv'
        df.to_csv(csv_file_path, index=False)

        print(f"Converted {file} to {csv_file_path}")
```
